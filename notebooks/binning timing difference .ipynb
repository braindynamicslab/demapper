```python
# This is not a fair comparison because its using different parameters for Resolution! 
# while one has bins, the other has number of landmarks

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

in_time = 34 * 60 + 29 # 34:29
ex_time = 1* 60* 60 + 44 * 60 + 13 # 1:44:13
cnt = 2000

table = [
    {
        'type': 'extrinsic',
        'time': ex_time / cnt
    },
    {
        'type': 'intrinsic',
        'time': in_time / cnt
    }
]

df = pd.DataFrame(data=table)
```

```python
sns.set(rc={'figure.facecolor':'white'})

g = sns.catplot(
    data=df, kind="bar",
    x="type", y="time",
    ci="sd", palette=sns.color_palette("dark")[1:], alpha=.6, height=6
)

g.despine(left=True)
g.set_axis_labels("", "Mappers per second")
# g.set(ylim=(0, 1))
# g.legend.set_title("")

plt.show()
```
